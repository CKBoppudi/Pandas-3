# Question 13 - Find out the most Productive director. (Director who have produced most number of movies per year).

```python
df.head(2)
```

```python
df.groupby(["director_name"])["year"].min()
```

```python
df.groupby(["director_name"])["year"].max()
```

```python
df.groupby(["director_name"])["title"].count()
```

```python
df_agg = df.groupby(["director_name"])[["year" , "title"]].aggregate({"title" : "count" , "year" : ["min" , "max"]})
```

```python
df_agg
```

```python
df_agg.columns
```

```python
df_agg.columns = ["_".join(col) for col in df_agg.columns]
```

```python
df_agg
```

```python
df_agg.reset_index(inplace = True)
```

```python
df_agg
```

```python
df_agg["years_active"] = df_agg["year_max"] - df_agg["year_min"]
```

```python
df_agg
```

```python
df_agg["movies_per_year"] = df_agg["title_count"] / df_agg["years_active"]
```

```python
df_agg
```

```python
df_agg.sort_values(["movies_per_year"] , ascending = False)
```

```python
df_agg.sort_values(["movies_per_year"] , inplace = True , ascending = False)
```

# Question 15 - Find out all the instances where the budget of a particular director for a movie is expensive. The movie is expensive when the budget of the particular movie is more than the average budget of the particular director.

```python
def budget_avg(x):
    x = x - x.mean()
    return x
df.groupby(["director_name"])["budget"].transform(budget_avg)
```
